```python
import re
import os
import pandas as pd
from bs4 import BeautifulSoup, CData
```

```python
data_df = pd.DataFrame(columns=['billNumber','originChamber','billType','introducedDate',
                                'congress','title','xml_url','summary','bioguideId',
                                'fullName','firstName','middleName','lastName','party','state'])
```

```python
file_name = 'govtrack_data'
```

```python
def get_infor(soup):
    infor_list = []
    '''议案基本信息'''
    infor_list.append(soup.bill.billNumber.string)
    infor_list.append(soup.bill.originChamber.string)
    infor_list.append(soup.bill.billType.string)
    infor_list.append(soup.bill.introducedDate.string)
    infor_list.append(soup.bill.congress.string)
    infor_list.append(soup.title.string)
    try:
        infor_list.append(soup.textVersions.item.formats.item.url.string)
    except AttributeError:
        infor_list.append('None')
    try:
        summary = re.sub(r'<.*?>','',re.search('<p>(.*)</p>', soup.summaries.item.text).group())
        infor_list.append(summary)
    except AttributeError:
        summary = soup.title.string
        infor_list.append(summary)
        
    '''资助者信息'''
    infor_list.append(soup.sponsors.item.bioguideId.string)
    infor_list.append(soup.sponsors.item.fullName.string)
    infor_list.append(soup.sponsors.item.firstName.string)
    infor_list.append(soup.sponsors.item.middleName.string)
    infor_list.append(soup.sponsors.item.lastName.string)
    infor_list.append(soup.sponsors.item.party.string)
    infor_list.append(soup.sponsors.item.state.string)
    return infor_list
```

```python
for root, dirs, files in os.walk(file_name):
    for f in files:
        if str(f) == '.DS_Store':
            continue
        print(os.path.join(root, f))
        soup = BeautifulSoup(open(os.path.join(root, f)),'xml')
        infor_list = get_infor(soup)
        data_df.loc[len(data_df)] = infor_list
```
